```python
df['normalize_km'] = (df['km'] - df['km'].min()) / (df['km'].max() - df['km'].min())
```

```python
def predict_price(km: pd.core.series.Series):
    return b0 + b1 * km
```

```python
b0 = 0
b1 = 0
df['predicted_price'] = predict_price(df['normalize_km'])
df['residual'] = df['price'] - df['predicted_price']
```
